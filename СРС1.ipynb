```
from google.colab import files
files.upload()

import random
import time
import matplotlib.pyplot as plt
from реализация_алгоритмов_срс1 import insertion_sort, quick_sort
```

```
#cлучайный массив
def random_array(n):
    return [random.randint(1, 10000) for _ in range(n)]

#отсортированный массив
def sorted_array(n):
    return list(range(1, n + 1))

#массив в обратном порядке
def reversed_array(n):
    return list(range(n, 0, -1))

#почти отсортированный массив
def almost_sorted_array(n, disorder=0.05):

    arr = list(range(1, n + 1))

    num_swaps = max(1, int(n * disorder))

    for _ in range(num_swaps):
        i, j = random.sample(range(n), 2)
        arr[i], arr[j] = arr[j], arr[i]

    return arr

sizes = [1000, 5000, 10000, 50000]
```

```
def test (array, name):

    times_insertionsort = []
    times_quicksort = []


    for n in sizes:

        arr1 = array(n)
        arr2 = arr.copy()

        # измерение для первого алгоритма
        start = time.time()
        insertion_sort(arr1)
        end = time.time()
        times_insertionsort.append(end - start)

        # измерение для второго алгоритма
        start = time.time()
        quick_sort(arr2, 0, len(arr2) - 1)
        end = time.time()
        times_quicksort.append(end - start)


    # построение графика
    plt.plot(sizes, times_insertionsort, marker='o', label='Insertion Sort function O(n²)')
    plt.plot(sizes, times_quicksort, marker='o', label='QuickSort function O(n(logn))')

    plt.xlabel("Размер входных данных (n)")
    plt.ylabel("Время выполнения (секунды)")
    plt.title(name)

    plt.legend()
    plt.grid(True)

    plt.show()
```

```
test(random_array, "Случайный массив")
test(sorted_array, "Отсортированный массив")
test(reversed_array, "Обратно отсортированный массив")
test(almost_sorted_array, "Почти отсортированный массив")
```
