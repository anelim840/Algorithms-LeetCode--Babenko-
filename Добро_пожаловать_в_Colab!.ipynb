```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
from google.colab import files
files.upload()

import random
import time
import matplotlib.pyplot as plt
from реализация_алгоритмов_срс1 import insertion_sort, quick_sort
```

```
# Случайный массив
def generate_random(n):
    return [random.randint(1, 10000) for _ in range(n)]

# Отсортированный массив
def generate_sorted(n):
    return list(range(1, n + 1))

# Массив в обратном порядке
def generate_reverse(n):
    return list(range(n, 0, -1))

# Почти отсортированный массив
def generate_nearly_sorted(n, disorder=0.05):

    arr = list(range(1, n + 1))

    num_swaps = max(1, int(n * disorder))

    for _ in range(num_swaps):
        i, j = random.sample(range(n), 2)
        arr[i], arr[j] = arr[j], arr[i]

    return arr
```

```
sizes = [1000, 5000, 10000, 50000]
```

```
def test (array, name):

    times_insertionsort = []
    times_quicksort = []


    for n in sizes:

        arr = array(n)
        arr2 = arr.copy()

        # измерение для первого алгоритма
        start = time.time()
        insertion_sort(arr)
        end = time.time()
        times_insertionsort.append(end - start)

        # измерение для второго алгоритма
        start = time.time()
        quick_sort(arr2, 0, len(arr2) - 1)
        end = time.time()
        times_quicksort.append(end - start)


    # построение графика
    plt.plot(sizes, times_insertionsort, marker='o', label='Insertion Sort function O(n²)')
    plt.plot(sizes, times_quicksort, marker='o', label='QuickSort function O(n(logn))')

    plt.xlabel("Размер входных данных (n)")
    plt.ylabel("Время выполнения (секунды)")
    plt.title(name)

    plt.legend()
    plt.grid(True)

    plt.show()
```

```
test(generate_random, "Случайный массив")
test(generate_sorted, "Отсортированный массив")
test(generate_reverse, "Обратно отсортированный массив")
test(generate_nearly_sorted, "Почти отсортированный массив")
```
